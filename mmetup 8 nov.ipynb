```python
import os
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
import statsmodels.api as sm
import numpy as np
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 16, 8
```

```python
#mydata = quandl.get("EOD/AAPL",start_date="2012-01-01",api_key="")
mydata=pd.read_csv("/home/gaetano/jupyter_test/meetUp_PyTrade/WIKI-AAPL.csv")
eoddata = mydata['Adj. Close']
eoddataV = eoddata.values
rtrns = np.log(eoddataV[1:]/eoddataV[0:-1])
print(np.mean(rtrns))
print(np.std(rtrns))
fig1=plt.figure(1)
plt.plot(rtrns)
plt.show()
#fig1.savefig("graph.png")
xvals = np.sort(rtrns)*100
yvals = np.arange(len(xvals))/float(len(xvals))*100
fig2=plt.figure(2)
plt.plot(xvals, yvals)
plt.xlabel('Daily Returns [%]')
plt.ylabel('CDF')
plt.show()
#fig2.savefig("graph2.png")
```

```python
#setting the Dates
mask = (mydata['Date'] > '2010-01-01')
dts=dts.loc[mask]

#setting the data for the 200MA
eoddataInverted = eoddata.reindex(index=eoddata.index[::-1])
eoddata200MA=eoddataInverted.rolling(window=200).mean()
eoddata200V=eoddata.values
eoddata200VMA=eoddata200MA.values

#drawing the two lines of the plot
ts=pd.Series(eoddataV,index=dts)
ts=ts.reindex(index=ts.index[::-1])
ts.plot()

tsMA=pd.Series(eoddata200VMA,index=dts)
tsMA.plot(color = 'g')
```
